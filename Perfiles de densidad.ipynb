# Mass Density Profiles


### Dark Matter Halo: Navarro Frenk White (NFW) profile ###
The Navarro–Frenk–White (NFW) profile is a spatial mass distribution of dark matter fitted to dark matter halos identified in N-body simulations by Julio Navarro, Carlos Frenk and Simon White. The NFW profile is one of the most commonly used model profiles for dark matter halos.

In the NFW profile, the density of dark matter as a function of radius is given by:

$$
\rho (r) = \frac{\rho_0}{\frac{r}{R_s} \left ( 1+ \frac{r}{R_s} \right )^{2}}	
$$

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [7, 5]

#linearly spaced numbers
r = np.arange(1e-05, 100)

# NFW profile
#**************

# Free Variables
rho_0 = 8.5e6 #Ms/kpc^3 --> GalactICs
R_s = 19.6 #kpc --> GalactICs

#NFW basic formula
rho = rho_0/((r/R_s)*(1 + r/R_s)**2)

# setting the axes and sublot
fig, ax = plt.subplots()

# plot the function
plt.loglog(r,rho)

# Add title and axis names
plt.title('NFW Profile')
plt.xlabel('Log R (kpc)')
plt.ylabel(r'Log $\rho$ ($M_{\odot}$ $kpc^-3$)')


# show the plot
plt.show()
```

### Dark Matter Halo: Einasto Profile ###

The Einasto profile (or Einasto model) is a mathematical function that describes how the density 
density ρ of a spherical stellar system varies with distance r from its center.
The Einasto profile (Einasto 1965) is defined by the density function:


$$
\rho (r) = \rho_0 * exp \left (  \frac{-2}{\alpha }  \left [ \left ( \frac{r}{r_s} \right )^{\alpha } -1 \right ]\right )
$$

```python

# NFW profile
#**************

# Free Variables
rho_0 = 8.5e6 #Ms/kpc^3 --> GalactICs
R_s = 19.6 #kpc --> GalactICs

#NFW basic formula
rho = rho_0/((r/R_s)*(1 + r/R_s)**2)

#Einasto approximation
#*********************
alpha = 0.15
rho1 = rho_0 * np.exp((-2/alpha)*(pow((r/R_s),alpha)-1))

# setting the axes and sublot
fig, ax = plt.subplots()

# plot the function
plt.loglog(r,rho)

# plot the function
plt.loglog(r,rho1)


# Add title and axis names
plt.title('NFW & Einasto Profiles')
plt.xlabel('Log R (kpc)')
plt.ylabel(r'Log $\rho$ ($M_{\odot}$ $kpc^-3$)')
plt.legend(["NFW", "Einasto"])


# show the plot
plt.show()
```

### Bulge: Sérsic Profile ###

The bulges of late-type galaxies are found to have surface brightness profiles which follow the more general Sérsic law.
In McMillan's paper* was choosen the target density profile for the bulge to be:


$$
p_b(r) = \frac{p_{0,b}}{\left ( 1+\frac{r'}{r_0} \right )^\alpha}  exp\left [  -\left ( \frac{r'}{r_{cut}} \right )^2\right ]
$$

with:
$$
r' = \sqrt{r + \left ( z/q \right )^2}
$$

```python
# Bulge density profile
#**************

# Free Variables
alpha_ = 1.8 #--> McMillan
r_0 = 0.075 #kpc --> McMillan
r_cut = 2.1 #kpc --> McMillan
q= 0.5 #--> McMillan
rho_b0 = 9.93e10 #Ms/kpc^3--> McMillan
z= 1 #z scale

r_ = np.sqrt(r+(z/q)**2)

rho_b = (rho_b0/(1+(r_/r_0))**alpha_) * np.exp(-(r_/r_cut)**2)

# setting the axes and sublot
fig, ax = plt.subplots()

# plot the function
plt.loglog(r,rho_b)

# Add title and axis names
plt.title('Sérsic Profile')
plt.xlabel('Log R (kpc)')
plt.ylabel(r'Log $\rho$ ($M_{\odot}$ $kpc^-3$)')

# show the plot
plt.show()
```
